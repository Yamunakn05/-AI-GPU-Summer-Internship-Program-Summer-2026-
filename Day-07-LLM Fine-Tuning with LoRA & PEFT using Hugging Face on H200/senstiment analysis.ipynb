```python
from groq import Groq

client = Groq(
    api_key="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"
)

prompt = """
Classify sentiment.

Example 1:
Review: The product is amazing.
Sentiment: Positive

Example 2:
Review: The service was terrible.
Sentiment: Negative

Example 3:
Review: The experience was okay.
Sentiment: Neutral

Now classify:

Review: The laptop performance is excellent and battery life is amazing.

Output only:
Positive, Negative, or Neutral
"""

response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ],
    temperature=0,
    max_tokens=10
)

print(response.choices[0].message.content)
```
