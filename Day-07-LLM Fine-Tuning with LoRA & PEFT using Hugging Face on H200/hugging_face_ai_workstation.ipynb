```python
import torch
import time
import psutil

from transformers import pipeline, AutoTokenizer, AutoModel

print("="*60)
print("HUGGING FACE AI WORKSTATION (100% COMPATIBLE)")
print("="*60)

# ---------------------------------------------------
# DEVICE
# ---------------------------------------------------

device = "cuda" if torch.cuda.is_available() else "cpu"
print("\nDevice:", device)

if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))

# ---------------------------------------------------
# TEXT
# ---------------------------------------------------

text = """
Artificial Intelligence is transforming education and cybersecurity.
"""

# ---------------------------------------------------
# TOKENIZATION
# ---------------------------------------------------

print("\nTOKENIZATION")

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

tokens = tokenizer.tokenize(text)

print(tokens)

# ---------------------------------------------------
# EMBEDDINGS
# ---------------------------------------------------

print("\nEMBEDDINGS")

model = AutoModel.from_pretrained("bert-base-uncased")

inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)

with torch.no_grad():
    outputs = model(**inputs)

print("Embedding shape:", outputs.last_hidden_state.shape)

# ---------------------------------------------------
# TEXT GENERATION
# ---------------------------------------------------

print("\nTEXT GENERATION")

generator = pipeline("text-generation", model="distilgpt2")

print(generator("Machine Learning is", max_new_tokens=40)[0]["generated_text"])

# ---------------------------------------------------
# SENTIMENT ANALYSIS
# ---------------------------------------------------

print("\nSENTIMENT")

sentiment = pipeline("sentiment-analysis")

print(sentiment("This workshop is amazing"))

# ---------------------------------------------------
# SUMMARIZATION (SAFE VERSION)
# ---------------------------------------------------

print("\nSUMMARIZATION")

summary_prompt = f"""
Summarize this in 2 lines:

{text}

Summary:
"""

summary = generator(summary_prompt, max_new_tokens=60)

print(summary[0]["generated_text"])

# ---------------------------------------------------
# TRANSLATION (FIXED - NO PIPELINE TASK)
# ---------------------------------------------------

print("\nTRANSLATION (WORKING VERSION)")

translation_prompt = """
Translate to French:
Artificial Intelligence is transforming education.

French:
"""

translation = generator(translation_prompt, max_new_tokens=40)

print(translation[0]["generated_text"])

# ---------------------------------------------------
# MEMORY
# ---------------------------------------------------

print("\nMEMORY")

process = psutil.Process()

print("RAM:", round(process.memory_info().rss / 1024**2, 2), "MB")

print("\nDONE")
```
